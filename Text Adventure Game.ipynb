```python
# Alice's Adventures in Wonderland Adventure Game
# Author: Camilla Bravi
# Date of Assignment: 10.25.21

"""

DocString:

A) Introduction:
   Welcome to the Alice's in Wonderland adventure game. This game refers to the
   novel by Lewis Carroll and is composed by four worlds. Jump from one to another
   and help Alice going back to the real world
   
B) Known Bugs/Errors:
   None
   
"""

def the_rabbit_hole():
    
    print("""WORLD 1 - THE RABBIT HOLE

Alice is in the garden when a white rabbit dressed in a jacket and waistcoat, 
passes by saying 'Poor me! Poor me! I'm running late' and looks worried at his 
watch. The rabbit enters a den and Alice follows him, finding herself in a dark 
gallery and then falling into a well, furnished with shelves and cupboards with 
jars.

""")
    print("Press 1 to go to the tea party.")
    
# Start of nested conditional 1
    
    choice = input("> ")
    
    if choice == '1':
        print("""Answer this question to proceed. 
        
Alice lands and finds herself in a room with a table and a tiny door from which
it is impossible to pass ... if only Alice could shrink! On the table there's 
a bottle with a tag that says 'DRINK ME'. Decide the next step to do!

1) wait for the white rabbit to come back
2) drink from the bottle
        
""")
        
# Start of nested conditional 2

        choice = input("> ")
                
        if choice == '1':
            print("That's incorrect.")
            fail()
 
           
        elif choice == '2':
            print("That's correct! Please enjoy the tea party!")
            input('<Press any key to continue>\n')
            the_tea_party()        
            
            
        else:
            print("Invalid entry. Please try again.\n")
            input('<Press any key to continue>\n')
            the_rabbit_hole()
            
# End of nested conditional 2        

    else:
        print("Invalid entry. Please try again.\n")
        input('<Press any key to continue>\n')
        the_rabbit_hole()
        
    
# End of nested conditional 1

def the_tea_party():
    
    print("""WORLD 2 - THE TEA PARTY
    
Alice arrives in front of a house and sits to a table, where a hare with a mad 
hatter and a dormouse are getting tea. They're going from one cup to another, 
completing the tour of all the seats, to taste every tea.

Alice asks them to help her get home but they keep interrupting her, urging 
her to keep doing the seats' tour. Should you stay or should you go?

1) stand up and leave
2) keep doing the tour of the seats
""")

    
# While Loop

    while True:  
        print("Guess the answer to meet the Queen of Hearts.")
        
        choice = input("> ")
    
        if choice == '1':
            the_queen_of_hearts()
            break
            
        elif choice == '2':
            print("Sure you want keep drinking tea??")
            
        else:
            print("Invalid entry. Please try again.\n")
            
            
def the_queen_of_hearts():
    
    print("""WORLD 3 - THE QUEEN OF HEARTS
        
Alice finds a path that leads her to a castle, where a Queen of Hearts is
playing golf in the garden. If you want to bring Alice back to the real 
world you have to answer this last question. 

In playing cards, which number does the Queen correspond to? Is it seven or nine?

""")
    choice = input("> ").lower().strip()
    
    if choice == 'seven' or choice == '7':
        print("That's incorrect.")
        fail()
            
    elif choice == 'nine'or choice == '9':
        print("That's correct! Please enjoy the real world!")
        input('<Press any key to continue>\n')
        the_real_world()
            
    else:
        print("Invalid entry. Please try again.\n")
        input('<Press any key to continue>\n')
        the_queen_of_hearts()


def the_real_world():
    
    print(f"""
    
Hey Hey {user_name}, you made it! Alice is back in the real world! Press any
number to win\n""")

    
# Control of errors with try/except function

    try:
        number = int(input("Enter a number: "))
        print(number)
        win()
        
    except ValueError:
        print("Invalid Entry")
        the_real_world()
        

# Definition of fail function

def fail():
    
    print("Oh no! You failed, but thank you for playing!")
    play_again = input('<Press 1 if you want to play again>\n')
    if play_again == '1':
        the_rabbit_hole()
    else:
        print("Thank you for playing")

        
# Definition of win function

def win():
    
    print("Congratulations! You win!")
    
           
##############################################################################

# Use of < prompt > with the triple-quote wrapper

user_name=input(prompt=""" 
Welcome to the crazy world of Alice's Adventures in Wonderland! What's your name?
""")
print(f"""Hi {user_name}! Let the adventure begin!""")

the_rabbit_hole()
```
